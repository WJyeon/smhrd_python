### 조건문

- if 라는 키워드로 시작
- if 다음에 조건식 작성
- if 조건식 :
- (들여쓰기) 실행문장
- 주의사항
    - 조건식에는 True/False 나오는 논리형의 결과를 써줘야 한다.
    - 반드시 콜론: !!
    - 반드시 들여쓰기!! -> 조건문의 실행영역을 구분
    

```python
# 조건식이 True일 경우
if True :
    print('실행문장 실행')
```

```python
# 조건식이 False일 경우
if False :
    print('실행문장 실행')
# 조건문의 실행영역에 들어올 수 없어서 아무 실행도 하지 않고 끝난다
```

```python
# 들여쓰기의 중요성
if True :
    print("실행문장 실행")
print('if문 바깥쪽에 있는 문장')
```

```python
# 들여쓰기의 중요성
if False :
    print("실행문장 실행")
print('if문 바깥쪽에 있는 문장')

#들여쓰기가 되어있지 않으면 if의 영향을 받지 않는다
```

```python
# 예제
money = int(input('money = '))

if money >= 10000 : 
    print('택시를 탄다')
```

### if else문
- if 조건식 : 
- (들여쓰기) 실행문장1
- else :
- (들여쓰기) 실행문장2
- 주의사항
    - if, else 같은 선 상에 있어야 한다!
    - 콜론, 들여쓰기
    - else는 참이 아니였을 모든 경우 (False)를 포함하기 때문에 조건식을 적지 않는다!
    


```python
if True:
    print('실행문장1')
else :
    print('실행문장2')
```

```python
if False:
    print('실행문장1')
else :
    print('실행문장2')

# 단순 if문과의 차이점: 조건문의 False였을 경우 실행문장 유무 차이
```

```python
# 예제
money = int(input('소지금액 = '))

if money >= 10000 : 
    print('택시를 탄다😊') #window+. 이모지 출력
else :
    print('버스를 탄다')
```

```python
# 예제2
num = int(input('정수 입력 >> '))
if (num%3) == 0 and (num%5) == 0 : #3과 5의 최소공배수 15를 사용해도 무방
    print("3과 5의 배수 입니다.")
else :
    print("3과 5의 배수가 아닙니다.")
```

```python
#마트 계산 프로그램
count = int(input('상품의 개수를 입력하세요 >> '))
price = 10000
sum1 = count * price

if count >= 11 :
    print(f'가격은 {sum1 - int(sum1*0.1)}입니다.')
else :
    print(f'가격은 {sum1}입니다.')
```

```python
# 마트 계산 프로그램 답
# 1. 상품의 개수 입력 받기
cnt = int(input('사려는 상품의 개수를 입력하세요 >> '))
# 2. 금액 계산 11개 미만이면~ 11개 이상이면!
if cnt >= 11 :
    print(f'가격은 {int(cnt*10000*0.9)}원 입니다.') #cnt*9000
else :
    print(f'가격은 {cnt*10000}원 입니다.')
```

### if elif, else문
 - elif문은 여러 개의 조건을 사용할 때 쓰는 키워드
 - if와 같은 선 상에 위치해야 함
 - else문은 반드시 작성하지 않아도 된다.



```python
#돈이 10000원 이상일 경우 택시, 10000미만 1000이상일 경우 버스, 1000미만일 경우 걸어간다
money = int(input('돈이 얼마나 있나요? >>'))
if money >= 10000 :
    print('택시를 탄다')
elif money >= 1000 :
    print('버스를 탄다')
else :
    print('걸어간다')
```

```python
num1 = int(input('첫 번째 정수 입력>> '))
num2 = int(input('두 번째 정수 입력>> '))

if num1 > num2 :
    print('첫번째 정수가 더 큽니다.')
elif num1 < num2 :
    print('두번째 정수가 더 큽니다.')
else :
    print('두 수가 같습니다.')
```

```python
score = int(input('점수 입력 >>'))

if score<= 100 and score >= 90 :
    print(f'{score}점은 A학점 입니다.')
elif score < 90 and score >= 80 :
    print(f'{score}점은 B학점 입니다.')
elif score < 80 and score >= 70:
    print(f'{score}점은 C학점 입니다.')
elif score < 70 and score >= 60:
    print(f'{score}점은 D학점 입니다.')
elif score >= 0 and score < 60:
    print(f'{score}점은 F학점 입니다.')
else :
    print('잘못된 점수입니다.')
```

```python
# 학점을 판별하는 조건문 작성 (5개의 조건문)

score = int(input("점수 입력 >> "))
grade = "-1"
if (score > 100 or score < 0):
    print("잘못된 점수 입력입니다.")
elif (score >= 90):
    grade = "A"
elif (score >= 80):
    grade = "B"
elif (score >= 70):
    grade = "C"
elif (score >= 60):
    grade = "D"
else:
    grade = "F"

if (grade != "-1"):
    print(f"{score}점은 {grade}학점입니다.")
```
