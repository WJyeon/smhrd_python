```python
score = int(input('  >>'))

if score<= 100 and score >= 90 :
    print(f'{score} A .')
elif score < 90 and score >= 80 :
    print(f'{score} B .')
elif score < 80 and score >= 70:
    print(f'{score} C .')
elif score < 70 and score >= 60:
    print(f'{score} D .')
elif score >= 0 and score < 60:
    print(f'{score} F .')
else :
    print(' .')
```

```python
#     (5 )

score = int(input("  >> "))
grade = "-1"
if (score > 100 or score < 0):
    print("  .")
elif (score >= 90):
    grade = "A"
elif (score >= 80):
    grade = "B"
elif (score >= 70):
    grade = "C"
elif (score >= 60):
    grade = "D"
else:
    grade = "F"

if (grade != "-1"):
    print(f"{score} {grade}.")
```
