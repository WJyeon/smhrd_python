### 대입연산자

 - +=, -+, *=, /=, %=(복합 대입 연산자) : 연산 후의 값을 앞에 대입한다 # 등호가 뒤

### 비교 연산자


 - 같다 == : 등호를 두개 사용함 주의!
 - 다르다 != : 
 - <, > <=, >= : 꺾쇠가 열린 쪽이 큰 값
 - 결과값은 항상 불리언형(논리형, True/ False)로 나온다

```python
a = 3
b = 4
print(a < b)
print(a == b)
print(a != b)
```

### 논리 연산자
 - 논리형(True, False) 결과값을 연산해주는 연산자
     - and : 두 개의 값이 모두 True일 때만 True 결과 도출
     - or : 둘 중 하나라도 True면 True 도출
     - not : 논리를 부정하는 연산자(논리형 앞에 써주기)

```python
a = 10
b = 7
print(a != b and a == b)
print(a != b or a == b ,'\n')

print(a > b)
print(not a > b)
```

### 삼항연산자
 - 조건식의 결과가 참이라면 a 실행, 거짓이면 b를 실행
 - 주의! 조건식의 결과는 항상 논리형(True, False)이어야 한다

```python
# 나이가 20살 이상이면 '성인', 20살 미만이면 '미성년자'출력
age = int(input('나이 입력: '))
print('성인')if age>=20 else print('미성년자')
```

```python
a = int(input("정수입력>> "))
b = int(input("정수입력>> "))

print(f'a : {a}') if a > b else print(f'b : {b}')
```

```python
a = int(input("첫번째 정수입력>> "))
b = int(input("두번째 정수입력>> "))

print(f'두 수의 차 : {a-b}') if a > b else print(f'두 수의 차 : {b-a}')
```
