# 반복문
- while : 주로 반복 횟수가 명확하지 않을 때
- for : 주로 반복 횟수가 정해져 있을 때

## while
~ 동안 : True인 동안 (False가 될 때 까지)

while 조건식:          ex. (num > 10) <br>
(tab) 실행 문장

```python
#반복문 사용해서 1~1000까지 출력 (while)

num=0
while (num <101):     
    num+=1
    print(num)
```

```python
#while사용해서 1~100
num=1
while True :
    print(num)
    num+=1
    if num >= 101: #if문을 사용한 반복문종료
        break      #반복문 빠져나가기
```

```python
#두개의 정수를 입력 받아 더하는 반복문
while True:
    num1 = int(input("첫번째 정수 입력 : "))
    num2 = int(input("두번째 정수 입력 : "))

    #num1과 num2가 0이 아니면 출력
    if num1 != 0 or num2 != 0 :
        print(f"두 정수의 합 : {num1+num2}입니다.")
    else:
        print("두 수 모두 0입니다")
        break
```

```python
cweight = int(input("현재 몸무게 : "))
pweight = int(input("목표 몸무게 : "))

week = 1
while cweight > pweight : #현재 몸무게가 목표 몸무게보다 큰 동안 반복
    rweight = int(input(f"{week}주차 감량 몸무게 : ")) #감량 무게
    cweight = cweight - rweight #감량한 현재 몸무게
    week +=1
print(f"{cweight}kg달성!! 축하합니다")
```

```python
current = int(input("현재 몸무게 "))
goal = int(input("목표 몸무게: "))

while True:
    minus = int(input("감량 몸무게: "))
    current = current - minus
    
    if current <= goal:
        print(f"{current}kg!! 다이어트 성공")
        break
```

```python
cor = r.randrange(1,51)
while True:
    num = int(input("숫자를 입력하세요 : "))
    if num == cor :
        print("정답을 맞추셨습니다.")
        break
    elif num > cor :
        print(f"{num}보다 작은 수 입니다.")
    else :
        print(f"{num}보다 큰 수 입니다.")
```
