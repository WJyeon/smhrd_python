### 문자열 인덱싱과 슬라이싱
- 문자열은 처음부터 끝까지 인덱스 번호를 배정받는다
    - 주의1. 인덱스는 0번부터 시작
    - 주의2. 띄어쓰기도 인덱스 번호를 가진다
- 인덱싱: 문자열 안에서 내가 원하는 데이터 하나를 뽑아올 때 사용
    - 인덱싱 방법: 변수명[인덱스 번호]
- 슬라이싱: 문자열 안에서 범위를 주고 잘라올 때 사용
    - 슬라이싱 방법: 변수명[시작 인덱스번호 : 끝 인덱스번호+1]

```python
name = 'My name is JY'
```

```python
#인덱싱 방법 : 변수명[인덱스번호]
#y 가져오기
name[1]
```

```python
#a 가져오기
name[4]
```

```python
#대문자 이니셜 첫글자 가져오기
#단순히 변수를 불러오는 것은 여러개를 불러도 마지막에 나온것만 보임. 다 보고 싶으면 print 사용
name[11]
name[-2]
```

```python
#My name 가져오기
#시작 인덱스 번호가 0 번이면 콜론 앞을 비워둬도 된다.
print(name[0:7])
print(name[:7]) # 콜론 앞 시작 인덱스가 비어있으면 '처음부터'라는 의미
```

```python
# i 가져오기 (인덱싱)
print(name[8], name[-5])

# name 가져오기 (슬라이싱)
print(name[3:7], name[-10:-6])

# initial가져오기 (슬라이싱)
print(name[-2:])

# 전체 다 가져오기 (슬라이싱)
print(name[:])
```

### 문자열 포매팅
- 문자열 안에 특정한 값(변수)을 바꿔야 할 때 사용할 수 있다.
- 3가지 방법
    - %기호 포매팅
    - format함수 포매팅
    - f문자열 포매팅

```python
# %기호 포매팅 -> 문자열 포맷 코드

month = 9
day = 27
s = '오늘은 %d월 %d일 입니다.'%(month,day) #순서대로 저장됨 ex) %(day, month)=> 27월 9일
print(s)
```

```python
#format함수 포매팅
s1 = '오늘은 {}월 {}일 입니다.'.format(month, day)
print(s1)
```

```python
# f문자열 포매팅
s2 = f'오늘은{month}월 {day}일 입니다.'
print(s2)
```

```python
x = 100
y = 200
sum2 = x + y
# %기호 포매팅
s1 = '%d와 %d의 합은 %d입니다.'%(x,y,sum2)
print(s1)

#format함수 포매팅
s2 = '{}와 {}의 합은 {}입니다.'.format(x,y,sum2)
print(s2)

#f문자열 포매팅
s3 = f'{x}와 {y}의 합은 {sum2}입니다.'
print(s3)
```
