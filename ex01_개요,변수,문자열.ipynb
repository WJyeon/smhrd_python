```python
# %  ->   

month = 9
day = 27
s = ' %d %d .'%(month,day) #  ex) %(day, month)=> 27 9
print(s)
```

```python
#format 
s1 = ' {} {} .'.format(month, day)
print(s1)
```

```python
# f 
s2 = f'{month} {day} .'
print(s2)
```

```python
x = 100
y = 200
sum2 = x + y
# % 
s1 = '%d %d  %d.'%(x,y,sum2)
print(s1)

#format 
s2 = '{} {}  {}.'.format(x,y,sum2)
print(s2)

#f 
s3 = f'{x} {y}  {sum2}.'
print(s3)
```
